We have a function that controls the flow of the game and a function that controls the brain of the bot. 

The main function 'PlayRPS', keeps track of the player's games played, games won, and their chosen move (Rock, Paper, or Scissors). The PlayRPS function, asks for the players choice and then asks the 'computer_brain' function get the computer's choice. To continue the game we created a while loop that stops once the player enters 'Q'. 

The 'computer_brain'function takes in the user's stats of choices and finds the most played option. If there are two or more options equally played, then the function randomly choses between the options through randit. 

After the computer's choice is made the winner is chosen by compairing the ints that represent Rock, Paper, or Scissors.This is accomplished by IF statements. 

Validation is wrapped in a Try-Except block to control wrong input's form the user.

```python
# import the random module
from random import randint

def computer_brain(user_stats):
    """Determines the optimal choice to beat the human in Rock Paper Scissors
    Returns: int
    Keyword Arguments:
    user_stats -- a dictionary containing the human's number of choices for rock, paper, and scissors
    """
    
    # Get the max value from the user's stats
    max_val = max(user_stats, key=user_stats.get)
    
    # determine which choice name is most used
    max_vals = [x for x, v in user_stats.items() if v == user_stats[max_val]]

    # If two options are the same make a random choice for the computer
    if len(max_vals) > 1:
        decision = randint(0, len(max_vals) -1)
    else:
        decision = 0
    
    # Make the choice from the most used moves
    choice = max_vals[decision]
    
    # Return the choice
    if choice == 'rock':
        return 1 
    elif choice == 'paper':
        return 2
    elif choice == 'scissors':
        return 0

def PlayRPS():
    """Starts and plays a game of Rock, Paper Scissors to challenge a human against a bot
    Returns:
    Keyword Arguments:
    """
    # Great the user and get their input
    print("Hello and welcome to Rock Paper Scissors.")
    user_input = str(raw_input("Enter 0 for Rock, 1 for Paper, 2 for Scissors, or Q to quit: "))
    print ''

    user_stats = {'rock': 0, 'paper': 0, 'scissors': 0 }
    game_stats = {'wins': 0, 'games': 0}
    options = ['Rock', 'Paper', 'Scissors']

    # loop through until the user enters 'Q'
    while user_input != 'Q' and user_input != 'q':
    
        # Validate the input corresponds to a Rock, Paper, Scissor Option
        try:
            if user_input not in ['0','1', '2']:
                raise Exception('You must enter 0, 1, 2 or Q')
        
            # Convert the user's input to an int
            user_choice = int(user_input)

            # Update the user's choice 
            user_stats[options[user_choice].lower()] += 1 

            # get the computer input
            computer_choice = computer_brain(user_stats)

            # Set the win variable to false until a user wins
            win = False

            if user_choice == computer_choice:
                
                print('It is a draw! You both played {}'.format(options[user_choice]))

            elif user_choice == 0:

                if computer_choice == 1:
                    print('You lose this round! Paper beats Rock')

                elif computer_choice == 2:
                    print('You win this round! Rock beats Sicssors')
                    win = True

            elif user_choice == 1:

                if computer_choice == 0:
                    print('You win this round! Paper beats Rock')
                    win = True

                elif computer_choice == 2:
                    print('You lose this round! Sicssors beats Paper')

            
            elif user_choice == 2:

                if computer_choice == 0:
                    print('You lose this round! Rock beats Sicssor')

                elif computer_choice == 1:
                    print('You win this round! Sicssors beats Paper')
                    win = True

            # Update the game stats
            game_stats['games'] += 1
            if win == True:
                game_stats["wins"] += 1
        
        except Exception as e:
            print str(e)
        
        finally:
            print ''
            user_input = raw_input("Enter 0 for Rock, 1 for Paper, 2 for Scissors, or Q to quit: ")

    # Print the output
    print 'Number of games played: {}'.format(game_stats['games'])
    print 'Number of game won {}'.format(game_stats['wins'])

PlayRPS()
```

We wanted to split user input and file processing into two functions. 

The 'GetMovieName' function handles the user input and validation. We thought a while loop with a try-except block would be easiest to continuously ask for the user's input if the validation of their reponses fails. 

The main function 'FindActors' controls the flow and file processing of the program. We have a for-loop that iterates through each line in the file, cleans the line and stores the results in a list that contains a tuple. The tuple stores the actors name, and another list that contains string of the movies the actor was in. 

We then loop through the filmography list to find which actors were in the user's desired movies. We store the actor's name in a Set corresponding to the movie.

Sets create the easiest way to determine if there are similarities between objects. So our actors for each movie are in a set. To find which actors are in the same movies we compare the sets and store the results and print the output. 

```python
# Get user input for movies
def GetMovieNames():
    """Gets the user's input for the two movies that they want to check for actors that are in both movies
    Return: Tuple of 2 strings -> Tuple(string, string)
    Arguement: None
    """
    
    # Ask for the names of the movies while the input is not valid
    valid_input = False
    while not valid_input:
        
        # Try to convert the movie names into string
        try:
            movie1 = str(raw_input('What is the name of movie 1: '))
            movie2 = str(raw_input('What is the name of movie 2: '))
            print ''
            
            # Validate that the movies are alnum strings
            no_space_movie1 = movie1.replace(" ", "")
            no_space_movie2 = movie2.replace(" ", "")

            if no_space_movie1.isalnum() and no_space_movie2.isalnum():
                valid_input = True
                return (movie1, movie2)
            else:
                raise Exception()
            
        except Exception as e:
            print 'The movies must be string'
            print ''

def FindActors():
    """Finds what actors are in the same movie
    Returns:
    Arguement:
    """

    # Open the file
    file = open('movies.csv', 'r')

    # Get the user's input
    movie1, movie2 = GetMovieNames()

    # Create an empty dictionary to hold the actors names and movie titles
    filmography = []

    # Loop through each line and extract the actor and movies
    for line in file.readlines():

        # Strip the line endings
        clean_line = line.rstrip()

        # Split the line in a list
        words = clean_line.split(',')

        # The first index is the actor's name
        actor_name = words[0]

        # Add the actor and their roles to a Tuple
        actor_roles = (actor_name, words[1:])
        filmography.append(actor_roles)
    
    # Sets to hold the names of the actors in each movie
    first_movie = set()
    second_movie = set()

    # Find the actors in each movie
    for actor in filmography:
        if movie1 in actor[1]:
            first_movie.add(actor[0])
        if movie2 in actor[1]:
            second_movie.add(actor[0])

    # Determine which actors are in the same movies
    both_movies = first_movie.intersection(second_movie)
    first_only = first_movie.difference(second_movie)
    second_only = second_movie.difference(first_movie)
    
    # Print the results
    print 'Actors in {} and {}: {}'.format(movie1, movie2, ', '.join(both_movies))
    print 'Actors only in the {}: {}'.format(movie1, ', '.join(first_only))
    print 'Actors only in the {}: {}'.format(movie2, ', '.join(second_only))

FindActors()
```

To build the rap song we thought only one function was needed. This keep the process clean and easy to understand. The function 'BuildSong' handles the flow and contruction of the song. 

We loop through the endings file to build two dictionaries that allow us to compare endings with rhytms and the beats needed to complete the rap phrase.

We are two dictionaries: 'endings' and 'rhythms' to make look ups easiers. 'endings' holds the ending words as the key and the value is a tuple containg the rhythm and the beats of the ending. 'rhythms' holds the 'rhythm' as the key and the value is a list of tuples. The list is of all the words that have the rhythm. The tuples contain the word of the same rhythm and the beats of the word. This two dictonary approach takes more memory but is faster than iteration. 

We then iterate through each line in the rap and match the correct endings. There are three possible situations for the lines handled by IF-ELSE statements.

The first situation is if the line needs an ending. Here we locate the words that have the same rhythm through the rhytms dictionary and then determine the beats needed from the previous lines total beats needed from the previous line and then change the ending of the line with the correct ending word.

The second situation is if the line needs no ending, in which case nothing changes. 

The third situation is if the line has an ending. Here we locate the ending words in the 'ending' dictionary and then get the total number of beats and rhythm to compare to the next line.

Lastly, we print the phrase and the correct ending word(s).

```python
def BuildSong():
    """Builds a rap song by matching work endings to create cohesive beat
    Returns:
    Arguements:
    """

    s_file = open('skeleton_SamuraiShowdown.txt', 'r')
    e_file = open('endings.txt', 'r')

    # Holds the ending words as key
    # Tuple containing the rhythm word and beats in word as value
    endings = {}
    
    # Holds the rhythm words as key
    # List contains Tuples containing the ending words and beats in word as value
    rhythms = {}

    # load the ending.txt file into memory
    for line in e_file:

        # Clean the line
        clean_line = line.strip()

        # Split the line into its componet parts
        # idx 0 = end words
        # idx 1 = rhythm word
        # idx 2 = beats in word
        words = clean_line.split('::')
        words[2] = int(words[2])

        # add the endings to the dictionary
        endings[words[0]] = (words[1], words[2])
        if words[1] in rhythms:

            # Get the List of words and beats
            words_and_beats = rhythms[words[1]]

            # idx 0 = rhythm words
            # idx 1 = beats in word
            words_and_beats.append((words[0], words[2]))
        
        else:

            # Add a list containing Tuples of words and beats
            rhythms[words[1]] = [(words[0], words[2])]

    previous_total_beats = None
    previous_rhythm = None

    for line in s_file:

        # Clean the line endings
        clean_line = line.strip()

        # Split the line into its componet parts
        # idx 0 = prefix words
        # idx 1 = beats in prefix
        # idx 2 = ending words or 'XXX'
        words = clean_line.split('::')
        words[1] = int(words[1])

        if words[2] == 'XXX':

            # Locate the word(s) that have the same rhythms
            same_rhythm_words = rhythms[previous_rhythm]

            # Determine the number of beats needed
            beats_needed = previous_total_beats - words[1]

            # Find the right word with correct rhythm and beats needed
            for ending_word in same_rhythm_words:
                
                # Match the correct beats needed
                if beats_needed == ending_word[1]:
                    
                    # change the last word
                    words[2] = ending_word[0]
        elif words[2] == '':
            # do nothing
            pass
        else:
            # Locate the ending words in the ending dictionary
            verse_stats = endings[words[2]]
            
            # Get the rhythm and beat to compare to the next iteration (line)
            previous_total_beats = words[1] + verse_stats[1]
            previous_rhythm = verse_stats[0]

        # Print the results of the rap phrase
        if words[0] == '':
            print words[2]
        else:
            print words[0] + ' ' + words[2]

BuildSong()
```
